```python
pd.set_option('display.max_rows',10)
pd.set_option('display.max_columns',50)
customer=df.drop_duplicates(subset="customer_id")
customer=customer.reset_index()
customer=customer.drop(['index'],axis=1)
final_df = pd.DataFrame()
for i in range(len(customer)):
    df1=df[["customer_id","product_id"]]
    customer_for_product=df1[df1.customer_id==customer['customer_id'][i]]
    #xx.groupby(["customer_id"])["customer_id"].count().reset_index()
    customer_for_product=customer_for_product.drop_duplicates(subset="product_id")
    #aa=aa.groupby(["customer_id"])
    #aa.size().reset_index(name='counts')
    final_df_plus=customer_for_product.product_id.reset_index()
    idd=customer["customer_id"][i]
    final_df_plus["customer_id_"f'{idd}']=1
    final_df_plus=final_df_plus.drop(['index'],axis=1)
    t=customer_for_product.product_id.tolist()
    def string(n):
        return str(n)
    str_list=list(map(string, t))
    final_df_plus.index = Series(str_list)
    final_df_plusT=final_df_plus.T
    final_df_plusT=final_df_plusT.drop(['product_id'],axis=0)
    final_df_plusT
    final_df=pd.concat([final_df,final_df_plusT],axis=0)
final_df=final_df.fillna(0)
#final_df=final_df-0.5
final_df
```
