```python
#讀取raw data
import pymysql
import pandas as pd

try:
    db = pymysql.connect(host='192.168.5.50',port=3305 ,user='root', passwd='root', db='hi8')
    cursor = db.cursor(cursor=pymysql.cursors.DictCursor)
    print("已連線資料庫：", db)
except pymysql.err.OperationalError as oe:
    print("mysql連線失敗：", oe)
    print("結束程式！")
    
    
cursor = db.cursor()  
SQLCode=' SELECT  *'
SQLCode+=' FROM hi8.hashtagXcategory '
cursor.execute(SQLCode)
vidDataS=cursor.fetchall()
vidDataS
```

```python
title=cursor.description
title_list=[]
for i in range (len(title)):    
    #print((title[i])[0])
    title_list.append((title[i])[0])
df1=pd.DataFrame(vidDataS)
df1.columns=title_list
df1=df1.dropna(axis='index',how='all',subset=['category'])
df1.drop(df1.loc[df1['category']==''].index, inplace=True)
df1
```

```python
df1_count=df1.groupby(["hashtag", "category"]).size().reset_index(name="出現次數")
df1_count
```

```python
df1_count=df1_count.dropna(axis=0,how='any')
df1_count.drop(df1_count.loc[df1_count['hashtag']=='#'].index, inplace=True)
df1_count.drop(df1_count.loc[df1_count['出現次數']==1].index, inplace=True)
df1_count=df1_count[~ df1_count['hashtag'].str.contains("#?", regex=False)]
#df1_count=df1_count.drop(df1_count.loc[df1_count['出現次數']==1].index, inplace=True)
df1_count
```
